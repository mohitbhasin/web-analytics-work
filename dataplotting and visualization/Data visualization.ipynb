```python
import matplotlib.pyplot as plt
# within IPython, use, the magic, "%matplotlib inline" or "%pylab inline"
%pylab inline
# Let's import some related packages (numpy, pandas, etc)
import numpy as np
import pandas as pd
```

```python
'''
Created on 18-april-2015
@author: Mohit_Bhasin
'''
#data used is Nyc Social Media Usage.
#Data collected from https://catalog.data.gov/dataset/nyc-social-media-usage-555a2

'''I chose this data because it has data of followers for every month.
It can be used to compare different platforms on the basis of data provided.
I campared number of followers of facebook with twitter and the graph shows facebook
followers are increasing at slow rate than twitter'''

#Reading data from csv file into df
df = pd.read_csv('NYC_Social_Media_Usage.csv')
#creating empty lists to store data from csv file
facebook = []
twitter = []
f_date = []
t_date = []
#if platform in df is facebook and followers data is available then append data and date in the list
#if platform in df is twitter and followers data is available then append data and date in the list
for i in df.values:
    if i[1]=='Facebook' and i[4] > 0.0:
            facebook.append(i[4])
            f_date.append(pd.to_datetime(i[3]))
    if i[1]=='Twitter' and i[4] > 0.0:
        twitter.append(i[4])
        t_date.append(pd.to_datetime(i[3]))   
#plotting first 40 facebook and twitter followers using matplotlib.
plt.plot(f_date[:40],facebook[:40],'-r', label="Facebook",linewidth=4)
plt.plot(t_date[:40],twitter[:40], '-g', label="Twitter",linewidth=4)
plt.legend(loc='upper right')
plt.xlabel('Date Sampled')
plt.ylabel('Followers')
plt.grid(True)
plt.show()
```

```python
'''
Created on 18-april-2015
@author: Mohit_Bhasin
'''

#data used is Nyc Social Media Usage.
#Data collected from https://catalog.data.gov/dataset/nyc-social-media-usage-555a2

'''Here I have considered twittter,facebook,tumblr and linked-in.
compared all four platforms to see which platform has highest followers.
The result shows tumblr has highest followers of the four, with 34.5% followers.
twitter has 26.9%, linked-in 20.8% and facebook 17.9%.
'''

facebook = []
twitter = []
tumblr = []
link = []
f_avg = 0
tw_avg = 0
tb_avg = 0
link_avg = 0
for i in df.values:
    if i[1]=='Facebook' and i[4] > 0.0:
        facebook.append(i[4])
    if i[1]=='Twitter' and i[4] > 0.0:
        twitter.append(i[4])
    if i[1]=='Tumblr' and i[4] > 0.0:
        tumblr.append(i[4]) 
    if i[1]=='Linked-In' and i[4] > 0.0:
        link.append(i[4])
for i in range(100):
    f_avg = facebook[i] + f_avg
    tw_avg = twitter[i] + tw_avg
    tb_avg = tumblr[i] + tb_avg
    link_avg = link[i] + link_avg
f_avg = f_avg * 0.01    
tw_avg = tw_avg * 0.01   
tb_avg = tb_avg * 0.01   
link_avg = link_avg * 0.01   
labels = 'Facebook', 'Twitter', 'Tumblr', 'Linked-In'
sizes = [f_avg ,tw_avg, tb_avg, link_avg]
explode = (0, 0, 0.2, 0)
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%.1f%%', shadow=True)
plt.axis('equal');
```
